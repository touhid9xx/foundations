# Data Visualization Basics: Histograms, Boxplots, Scatterplots

# Welcome! In this lesson, you will learn how to create simple and powerful graphs.

- We will use real datasets to make concepts easy and practical.

- By the end, you will know how to explore and share data with clear charts.

```python
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset tips
tips = sns.load_dataset('tips')
# print shape
print("Dataset shape:", tips.shape)
# Display first few rows
print(tips.head())
```

# What is a Histogram?

- A histogram is a bar graph that shows how many data points fall within certain ranges.
- It is great for seeing patterns, like how spread out tips are or which values are most common.

```python
# Let us make a histogram of the 'tip' column.
plt.figure(figsize=(8,4))
sns.histplot(tips['tip'], bins=10, kde=True, color='skyblue', edgecolor='black')
plt.title('Histogram of Tip Amounts')
plt.xlabel('Tip Amount ($)')
plt.ylabel('Count')
plt.show()
```

# Why Are Histograms Useful?

Histograms help us spot patterns quickly, like whether tips are usually small, large, or somewhere in between.
They are helpful for finding odd or surprising values too.

```python
# Let us look at another column, the total bill amount.
sns.histplot(tips['total_bill'], bins=15, kde=True, color='lightgreen', edgecolor='black')
plt.title('Distribution of Total Bill Amount')
plt.xlabel('Total Bill ($)')
plt.ylabel('Count')
plt.show()
```

# What is a Boxplot?

A boxplot is a smart graph that summarizes data using minimum, lower quartile, median, upper quartile, and maximum.
It quickly shows if data is balanced, if there are outliers, or if most values are high or low.

```python
# Boxplot of tips to see summary statistics and spot outliers
plt.figure(figsize=(7,3))
sns.boxplot(x=tips['tip'], color='orange')
plt.title('Boxplot of Tips')
plt.xlabel('Tip Amount ($)')
plt.show()
```

```python
# Compare tips by day using boxplots
plt.figure(figsize=(8,4))
sns.boxplot(x="day", y="tip", data=tips, palette="Set2")
plt.title('Tips by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Tip Amount ($)')
plt.show()
```

# What is a Scatterplot?

- A scatterplot lets us see the relationship between two variables.
- If points go up together, that means as one value increases, the other tends to increase.
- Scatterplots are great to spot clusters, trends, or patterns.

```python
# Scatterplot of total bill vs tip
plt.figure(figsize=(8,5))
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='sex', style='time', palette='Set1')
plt.title('Tip Amount vs Total Bill')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip Amount ($)')
plt.legend(title='Sex / Meal Time')
plt.show()
```

```python
# Add a trend line with Seaborn's regplot
sns.regplot(x='total_bill', y='tip', data=tips, scatter_kws={'alpha':0.6}, line_kws={'color':'red'})
plt.title('Tip vs Total Bill with Trend Line')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip Amount ($)')
```

```python
# Handling Missing Data for Visualization

# Sometimes data has missing values.
# It is important to handle or fill these so charts do not break.
# Let us check for and fix missing values.

# Check for missing values
tips.isnull().sum()
```

```python
#fiill missing values in 'tip' column with the mean tip
tips['tip'].fillna(tips['tip'].mean(), inplace=True)
# copy the dataframe to a new variable
tips_filled = tips.copy()
#spot checks for nulls again
print(tips_filled.isnull().sum())
```

# Compare Distributions: Multiple Groups

- We can see differences in data by splitting into groups.
- Let us look at how tips differ by gender using histograms and boxplots.

```python
# Histogram: Tips by Gender
plt.figure(figsize=(8,4))
sns.histplot(data=tips, x='tip', hue='sex', bins=10, kde=True, palette='husl', alpha=0.7)
plt.title('Tip Amounts by Gender')
plt.xlabel('Tip Amount ($)')
plt.ylabel('Count')
plt.show()
```

```python
# Boxplot: Tips by Smoker Status
plt.figure(figsize=(7,3))
sns.boxplot(x='smoker', y='tip', data=tips, palette='cool')
plt.title('Tip Amounts by Smoker Status')
plt.xlabel('Smoker')
plt.ylabel('Tip Amount ($)')
plt.show()
```

# Customizing Plots: Style and Context

You can easily change the look of plots with Seaborn's built-in themes.
Let us see how a simple setting can make charts look different.

```python
# Set a new style for all plots in the notebook
sns.set_style('darkgrid')
plt.figure(figsize=(8,4))
sns.histplot(tips['tip'], bins=10, kde=True, color='purple')
plt.title('Histogram with Darkgrid Style')
plt.xlabel('Tip Amount ($)')
plt.show()
```

```python
# Mini-project: Explore all columns visually
for col in ['total_bill', 'tip', 'size']:
    plt.figure(figsize=(7,3))
    sns.histplot(tips[col], bins=10, color='steelblue', kde=True, edgecolor='black')
    plt.title(f"Distribution of {col.title()}")
    plt.xlabel(col.title())
    plt.ylabel('Count')
    plt.show()
```
